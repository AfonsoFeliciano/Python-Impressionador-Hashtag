# Mais edições de Gráfico com MatplotLib

```python
#importando o matplotlib
import matplotlib.pyplot as plt
```

- Usando o módulo numpy para trabalhar com números

```python
#importando o módulo numpy
import numpy as np
```

### Outros tipos de Gráfico:

- Linha

```python
vendas = np.random.randint(1000, 3000, 50)
meses = np.arange(1, 51)
```

```python
plt.plot(meses, vendas)
plt.axis([0, 50, 0, max(vendas) + 200])
plt.xlabel('Meses')
plt.ylabel('Vendas')
plt.show()
```

- Editando o Gráfico de Linha

```python
plt.plot(meses, vendas, '-')
plt.axis([0, 50, 0, max(vendas) + 200])
plt.xlabel('Meses')
plt.ylabel('Vendas')
plt.show()
```

```python
#editando o tamanho e a cor da linha - quais parâmetros editar?
plt.plot(meses, vendas, '-', color='orange')
plt.axis([0, 50, 0, max(vendas) + 200])
plt.xlabel('Meses')
plt.ylabel('Vendas')
plt.show()
```

- Dispersão

```python
plt.scatter(meses, vendas)
plt.show()
```

- Barras

```python
plt.bar(meses, vendas)
plt.show()
```

- Trabalhando com Múltiplos Gráficos no mesmo "Plot" -> Para melhor visualização/comparação

```python
plt.figure(1, figsize=(15, 3))
plt.subplot(1, 3, 1) #linha 1, 3 colunas, 1 = grafico à esquerda
plt.plot(meses, vendas, '-', color='orange')
plt.subplot(1, 3, 2) #linha 1, 3 colunas, 2 = grafico no meio
plt.scatter(meses, vendas)
plt.subplot(1, 3, 3) #linha 1, 3 colunas, 3 = grafico à direita
plt.bar(meses, vendas)
plt.show()
```
